```python
data.groupby(['Brand']).count()['Review #'].nlargest(10)
top_brand = list(data.groupby(['Brand']).count()['Review #'].nlargest(10).index)
df_agg = data.groupby(['Brand']).count()
df_agg.iloc[df_agg.index.isin(top_brand,level='Brand')].sort_values(by='Review #', ascending=False).sort_index(level='Brand',sort_remaining=False)
data[pd.notna(data['Top Ten'])]

def get_year(a):
    if len(a)<4:
        return np.NaN
    return int (a[:4])
def get_rank(a):
    if len(a)<4:
        return np.NaN
    return int (a[6:])

df_rank = data[pd.notna(data['Top Ten'])].reset_index()
df_rank['Tahun Rating'] = df_rank['Top Ten'].apply(lambda x:get_year(x))
df_rank['Rating'] = df_rank['Top Ten'].apply(lambda x:get_rank(x))
df_rank.dropna(inplace=True)
df_rank.sort_values(by=['Tahun Rating','Rating'],ascending=[False,True])

data.loc[data['Stars']=='Unrated','Stars']=0
data['Stars'] = data['Stars'].astype(float)
Brand_count = data[data['Stars']==5].groupby('Brand').count()['Stars'].nlargest(1).values[0]
Brand_name = data[data['Stars']==5].groupby('Brand').count()['Stars'].nlargest(1).index.values[0]

nissin_data = data[data['Brand']=='Nissin']
nissin_data['Stars']
nissin_data.boxplot(column=['Stars'], by='Country', fontsize=10, rot=45, figsize=(15,15), notch= True, patch_artist=True)
```

```python
def plot_top_brand(country):
    brand_count = data[data['Country']==country]['Brand'].value_counts()
    brand_count_30 = data[data['Country']==country]['Brand'].value_counts()[:30]
    fig = plt.figure(figsize=(15,10))
    ax = sns.barplot(x=brand_count_30.index,y=[count/sum(brand_count) for count in brand_count_30])
    ax.set_title("Merek Mie Ramen yang Paling Banyak Disukai di {}".format(country))
    ax.set_xticklabels(brand_count_30.index, rotation=45);
    
plot_top_brand('Indonesia')
```
